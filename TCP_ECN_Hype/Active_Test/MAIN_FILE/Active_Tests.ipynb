```python
from scapy.all import *
import random

def check_ecn_support(target_ip, target_port=80, output_file="ecn_response_ips_HTTP_PathSpider_with_ECT(1).txt"):
    sport = random.randint(1024, 65535)
    seq = random.randint(0, 4294967295)
    
    # Constructing the IP layer with ECT(1)
    ip = IP(dst=target_ip, tos=0x01)  # ECT(1) for the SYN packet
    
    # Constructing the TCP layer with SYN flag
    tcp = TCP(sport=sport, dport=target_port, flags='S', seq=seq)
    
    # Sending the SYN packet and waiting for a SYN-ACK
    syn_pkt = ip/tcp
    syn_ack_pkt = sr1(syn_pkt, timeout=2, verbose=0)
    
    if syn_ack_pkt and syn_ack_pkt.haslayer(TCP) and syn_ack_pkt.haslayer(IP):
        # Send an ACK packet to complete the three-way handshake
        ack_pkt = IP(dst=target_ip, tos=0x01)/TCP(sport=sport, dport=target_port, flags='A', seq=syn_ack_pkt[TCP].ack, ack=syn_ack_pkt[TCP].seq + 1)
        send(ack_pkt)
        
        # Sending an HTTP GET request
        get_request = "GET / HTTP/1.1\r\nHost: {}\r\n\r\n".format(target_ip)
        http_pkt = IP(dst=target_ip, tos=0x01)/TCP(sport=sport, dport=target_port, flags='PA', seq=syn_ack_pkt[TCP].ack, ack=syn_ack_pkt[TCP].seq + 1)/get_request
        http_response = sr1(http_pkt, timeout=2, verbose=0)
        
        # Checking the ECN bits in the IP header of the response
        if http_response and http_response.haslayer(IP):
            ecn_response = http_response[IP].tos & 0x03
            ecn_status = "Non-ECN" if ecn_response == 0 else "ECT(0)" if ecn_response == 2 else "ECT(1)" if ecn_response == 1 else "CE"
            
            with open(output_file, "a") as file:
                file.write(f"{target_ip} responded to HTTP GET with {ecn_status}\n")
        else:
            with open(output_file, "a") as file:
                file.write(f"{target_ip} did not respond to HTTP GET request\n")
    else:
        with open(output_file, "a") as file:
            file.write(f"No response or no ECN in IP header from {target_ip} during handshake\n")

# Replace the file path with your file containing IP addresses
with open('/Users/ertugrulgazitekden/Desktop/NALs/ECN_Transversial/Active_Test/Servers_List/ips1000k.txt', 'r') as file:
    for line in file:
        target_ip = line.strip()
        check_ecn_support(target_ip)
```

```python
from scapy.all import *
import random

def check_ecn_support(target_ip, target_port=80, output_file="ecn_Handshake_ips_2.txt", response_output="server_response_ips_2_IPLayer_After_TCP.txt"):
    sport = random.randint(1024, 65535)
    seq = random.randint(0, 4294967295)
    ip = IP(dst=target_ip, tos=0x02)  # ECT(0) in IP header for the SYN packet
    
    # Initial SYN packet with ECN flags set
    tcp = TCP(sport=sport, dport=target_port, flags='SEC', seq=seq)
    syn_pkt = ip/tcp
    syn_ack_pkt = sr1(syn_pkt, timeout=2, verbose=0)

    if syn_ack_pkt and syn_ack_pkt.haslayer(TCP) and syn_ack_pkt.haslayer(IP):
        # Check if ACK and ECE flags are set
        if syn_ack_pkt[TCP].flags & 0x52 == 0x52:  # ACK + ECE set
            with open(output_file, "a") as file:
                file.write(f"{target_ip} responded with ACK + ECN-Echo flags set\n")
            
            # Send another packet with sequence number incremented by 1 and ECT(0) set in the IP header
            ip_followup = IP(dst=target_ip, tos=0x02)  # ECT(0) set explicitly
            tcp_followup = TCP(sport=sport, dport=target_port, flags='A', seq=seq+1, ack=syn_ack_pkt[TCP].seq + 1)
            followup_pkt = ip_followup/tcp_followup
            followup_response = sr1(followup_pkt, timeout=2, verbose=0)
            
            # Log the server's response to the follow-up packet
            if followup_response:
                ecn_info = followup_response[IP].tos & 0x03
                with open(response_output, "a") as file:
                    file.write(f"{target_ip} responded to the follow-up packet with ECN: {ecn_info}\n")
            else:
                with open(response_output, "a") as file:
                    file.write(f"{target_ip} did not respond to the follow-up packet\n")
        else:
            with open(output_file, "a") as file:
                file.write(f"{target_ip} did not respond with ACK + ECN-Echo flags set\n")
    else:
        with open(output_file, "a") as file:
            file.write(f"No response from {target_ip}\n")

# Replace the file path with your file containing IP addresses
with open('/Users/ertugrulgazitekden/Desktop/NALs/ECN_Transversial/Active_Test/Servers_List/ips1000k.txt', 'r') as file:
    for line in file:
        target_ip = line.strip()
        check_ecn_support(target_ip)
```

```python
from scapy.all import *
import random

def check_ecn_support(target_ip, target_port=80, output_file="ecn_Handshake_Follow_up.txt", response_output="server_response_Http_Get_IPLayer_After_TCP.txt", http_response_output="http_responses.txt"):
    sport = random.randint(1024, 65535)
    seq = random.randint(0, 4294967295)
    ip = IP(dst=target_ip, tos=0x02)  # ECT(0) in IP header for the SYN packet
    
    # Initial SYN packet with ECN flags set
    tcp = TCP(sport=sport, dport=target_port, flags='SEC', seq=seq)
    syn_pkt = ip/tcp
    syn_ack_pkt = sr1(syn_pkt, timeout=2, verbose=0)

    if syn_ack_pkt and syn_ack_pkt.haslayer(TCP) and syn_ack_pkt.haslayer(IP):
        # Check if ACK and ECE flags are set
        if syn_ack_pkt[TCP].flags & 0x52 == 0x52:  # ACK + ECE set
            with open(output_file, "a") as file:
                file.write(f"{target_ip} responded with ACK + ECN-Echo flags set\n")
            
            # Construct an HTTP GET request as a follow-up
            get_request = 'GET / HTTP/1.1\r\nHost: {}\r\n\r\n'.format(target_ip)
            http_get_pkt = IP(dst=target_ip, tos=0x02)/TCP(sport=sport, dport=target_port, flags='A', seq=seq+1, ack=syn_ack_pkt[TCP].seq + 1)/Raw(load=get_request)
            http_response = sr1(http_get_pkt, timeout=2, verbose=0)

            # Save the HTTP response and check for ECT(0)
            if http_response and http_response.haslayer(IP):
                ect_value = http_response[IP].tos & 0x03  # Extract ECT bits
                if ect_value == 0x02:  # ECT(0) is set
                    ect_status = "ECT(0) is set in the response."
                else:
                    ect_status = "ECT(0) is not set in the response."

                with open(http_response_output, "a") as http_file:
                    if http_response.haslayer(Raw):
                        response_data = http_response[Raw].load.decode(errors='ignore')
                        http_file.write(f"Response from {target_ip}:\n{response_data}\n{ect_status}\n\n")
                    else:
                        http_file.write(f"{target_ip} responded without data. {ect_status}\n\n")
            else:
                with open(http_response_output, "a") as http_file:
                    http_file.write(f"{target_ip} did not respond to HTTP GET request\n")
        else:
            with open(output_file, "a") as file:
                file.write(f"{target_ip} did not respond with ACK + ECN-Echo flags set\n")
    else:
        with open(output_file, "a") as file:
            file.write(f"No response from {target_ip}\n")

# Replace the file path with your file containing IP addresses
with open('/Users/ertugrulgazitekden/Desktop/NALs/ECN_Transversial/Active_Test/Servers_List/ips1000k.txt', 'r') as file:
    for line in file:
        target_ip = line.strip()
        check_ecn_support(target_ip)
```
